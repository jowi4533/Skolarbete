```python
"""setFrames split videos to frames"""

i = 0
img_path = 'img'
video_path = 'videos'
training_data = []
total_frames = 0
def setFrames(the_id, gesture, i):
    total_frames = 0
    video_location = video_path+'/'+the_id+'/'+gesture+'.webm'
    
    cap = cv2.VideoCapture(video_location)
    counter = 0
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret == False:
            break
            
        if i % 30 == 0:
            counter+=1
            total_frames+=1
            img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            img = cv2.resize(img, (640, 480))
            data = X.iloc[i:i+1]
            
            print('i = ' + str(i))
            create_grid_boxes(img, data)
            
        i+=1
    cap.release()
    cv2.destroyAllWindows()
    print('No. of frames: ' + str(counter))
    return i
```

```python
""""this function split each image into a grid with 20x20px boxes and save the smaller gridimages with
    corresponding landmark-label to a big array.
"""


grid_image_array = []
all_landmark_counter = 0

def create_grid_boxes(img, data):

    px_x_mov = 32
    px_y_mov = 24
    x_max = 640
    y_max = 480
    y_pos = 0
    x_pos = 0
    land_mark_counter = 0
    actuallandmarkcounter = 0
    while y_pos < y_max:
        while x_pos < x_max:
            grid_img_label = "no_landmark"
            
            grid_img = img[y_pos:(y_pos+px_y_mov), x_pos:(x_pos+px_x_mov)]
            for i in range(0,(len(X.columns)),2):
                x_cord = data.iloc[0][i]
                y_cord = data.iloc[0][i+1]
                
                if (x_cord != 0) and (y_cord != 0) and (x_pos <= x_cord <= (x_pos+px_x_mov)) and (y_pos <= y_cord <= (y_pos+px_y_mov)):
                    land_mark_counter +=1

                    grid_img_label = data.columns[i]
                                            
            grid_image_array.append([grid_img, grid_img_label])
            x_pos += px_x_mov
            
        x_pos = 0
        y_pos += px_y_mov

    print('Landmark counter in this sub-img: ' + str(land_mark_counter))
```

```python
"""This function takes one array of images of a specific landmark and combinds it with no_landmark images.
This function is used to create data sets for training the models """

imgs = []
labels = []
def appendNoLandmark(landmark_array, no_landmarks_array):
    training_data = []
    interim_array = []
    complete_array = []
    number = 0
    imgs = []
    labels = []
    training_set = []
    label = []
    
    number = len(landmark_array)
    interim_array = random.sample(no_landmarks_array, number)
    
    complete_array = landmark_array + interim_array
    
    training_data = np.array(complete_array)

    np.random.shuffle(training_data)
    
    for img, label in training_data:
        imgs.append(img)
        labels.append(label)
    
    training_set = np.array(imgs)
    label = np.array(labels)
    
    return(training_set, label)
```

```python
""" Our Convolutional Neural Network"""
def createModel(training_set, label):
    training_set = np.array(training_set).reshape(-1,24,32,1)
    model = Sequential()
    model.add(Conv2D(5,(3,3), input_shape = (24,32,1))) #3x3 pixel window, 
    model.add(Activation('relu'))
    model.add(Flatten())
    model.add(tf.keras.layers.Dense(5))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    opt = tf.keras.optimizers.Adam(learning_rate=0.001)
    model.compile(loss="binary_crossentropy", optimizer=opt, metrics = ["accuracy"])
    model.fit(x=training_set, y=label, 
          batch_size = 50, 
          epochs = 150);
    return model
```

```python
"""this function predicts if, and which landmark that is present in a small grid image of size 20x20px
    If several models predict a specific landmark to be present in a grid image we choose the landmark with highest probability
"""

def predict_grid_array(img):
    px_x_mov = 32
    px_y_mov = 24
    x_max = 640
    y_max = 480
    y_pos = 0
    x_pos = 0
    land_mark_counter = 0
    all_pred_grid_img_arrs = []
    counter = 0
    
    while y_pos < y_max:
        while x_pos < x_max:
            grid_img = img[y_pos:(y_pos+px_y_mov), x_pos:(x_pos+px_x_mov)]
            grid_img = np.array(grid_img).reshape(1,24,32,1)
            grid_img_model_pred_val_arr = []
            for model,landmark in models:
                y_pred = model.predict(grid_img)
                #print(y_pred)
                #if we predict a landmark we append the prediction score
                if(model.predict_classes(grid_img) == 1):
                    grid_img_model_pred_val_arr.append([landmark,y_pred[0][0]]) 
                #else we just append 0.
                else:
                    grid_img_model_pred_val_arr.append(['no_landmark',0]) 
                     
            all_pred_grid_img_arrs.append(grid_img_model_pred_val_arr)
            x_pos += px_x_mov
            counter+=1
        x_pos = 0
        y_pos += px_y_mov
        print(counter)
        
    return(all_pred_grid_img_arrs) 
```

```python
"""We make the predictions of landmarks of a full image of size 640x480px """
def predict_all_grids(all_pred_arr):
    test_array = []
    used_landmarks = []
    #loops over each sub-image's array with 40 probabilities & landmark name
    for grid_pred_arr in all_pred_arr:
        highest_accuracy = ["no_landmark", 0]
        for landmark,pred_val in grid_pred_arr:
            pred_val = int(pred_val)
            if((pred_val > highest_accuracy[1]) and (landmark not in used_landmarks)):
                highest_accuracy[0] = landmark
                highest_accuracy[1] = pred_val
        test_array.append(highest_accuracy)
        used_landmarks.append(highest_accuracy[0])
    return(test_array)
```
